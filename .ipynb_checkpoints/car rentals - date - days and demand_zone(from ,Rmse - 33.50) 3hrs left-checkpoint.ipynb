```python
## demand_quarter

demand_quarter = []
for i in range(len(data['Quarter'])):
    if data['Quarter'][i] ==3:
        demand_quarter.append('Low')
    else:
        demand_quarter.append('High')
data['demand_quarter'] = demand_quarter
```

```python
demand_day = []
for i in range(len(data['day-of-week'])):
    if data['day-of-week'][i] <=3:
        demand_day.append("Low")
    else:
        demand_day.append("High")
```

```python
# hour basis
demand_zone = []
for i in range(len(data['hour'])):
    if data['hour'][i] < 6 or data['hour'][i]>=22:
        demand_zone.append("Low")
    elif 6<=data['hour'][i]<=12 or 18<=data['hour'][i]<=22 :
        demand_zone.append("Medium")
    else:
        demand_zone.append("High")
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
import scikitplot as skplt
```

```python
X
```

```python
from sklearn.metrics import mean_absolute_error
for i in range(0,1000):
    x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state = i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train) 
    pred_test = lr.predict(x_test)
    if round(r2_score(y_train,pred_train)*100,1) == round(r2_score(y_test,pred_test)*100,1):
        print("At random state",i,"model perform very well")
        print("At random state",i)
        print('Training accuracy score is',r2_score(y_train,pred_train)*100)
        print('Testing accuracy score is',r2_score(y_test,pred_test)*100)
        mse = mean_squared_error(y_test,pred_test)
        print('RMSE :', np.sqrt(mse))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rlf=RandomForestRegressor()
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=.10,random_state=927)
params = {'n_estimators':[100,125],
        'min_samples_split':[1,8],
        'min_samples_leaf':[1,5],
        'max_features': range(1,4),
        'min_impurity_decrease':np.arange(0.0,0.5),
        }
grd = GridSearchCV(rlf,param_grid = params)
grd.fit(x_train,y_train)
print('best_pram', grd.best_params_)

rf=grd.best_estimator_ #reinstantiating with best params

rf.fit(x_train,y_train)
pred = rf.predict(x_test)
print(r2_score(y_test,pred)*100)
print(mean_absolute_error(y_test,pred))
mse = mean_squared_error(y_test,pred)
print('RMSE :', np.sqrt(mse))
```
